```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
url_bike_hel='https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/Katariina/HelsinkiBikers_cleanDataset.csv'
url_bike_esp='https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/Katariina/EspooBikers_cleanDataset.csv'
url_bike_van='https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/Katariina/VantaaBikers_cleanDataset.csv'
```

```python
#helsinki 2014-1-1,2019-10-31;  
apps_hel = pd.read_csv(url_bike_hel)

#vantaa 2015-01-01,2016-12-31
apps_van = pd.read_csv(url_bike_van)

apps_hel
#espoo 2014-1-3, 2019-5-13,Thu,2018-06-14,23:00:00;
#apps_esp = pd.read_csv(url_bike_esp,index_col=[1,2])
```

```python
#apps_hel.index = apps_hel.index.map(' '.join)
#apps_hel.index=pd.to_datetime(apps_hel.index,errors='coerce')
```

```python
#Generate a dataframe for the Helsinki total cycling data on each timeframe
apps_hel['Hel_Sum']=apps_hel.iloc[:,3:].sum(axis=1)
apps_hel_sum=apps_hel[['Weekday','Date','Time','Hel_Sum']]
apps_hel_sum
```

```python
#Generate a dataframe for the Vantaa total cycling data on each timeframe
apps_van['Van_Sum']=apps_van.iloc[:,3:].sum(axis=1)
apps_van_sum=apps_van[['Weekday','Date','Time','Van_Sum']]
apps_van_sum
```

```python
# Helsinki and Vantaa cycling data between 2015 and 2016
apps_hel_1516=apps_hel_sum.loc[apps_hel_sum.Date.between('2015-01-01','2016-12-31')]
apps_van_1516=apps_van_sum.loc[apps_van_sum.Date.between('2015-01-01','2016-12-31')]
hel_van_1516=pd.merge(apps_hel_1516, apps_van_1516, how='outer', on=['Weekday','Date','Time'])
hel_van_1516
```

```python
#Helsinki and Vantaa cycling data by date

hel_van_1516_by_date=hel_van_1516.groupby('Date').sum()
hel_van_1516_by_date
```

```python
#Helsinki and Vantaa daily cycling data on 2015-2016
fig, ax = plt.subplots(figsize=(20,6))
line, = ax.plot(hel_van_1516_by_date.Hel_Sum.sort_index(),label='Helsinki Cycling by date')
hel_by_date_max=hel_van_1516_by_date.Hel_Sum.max()
line, = ax.plot(hel_van_1516_by_date.Van_Sum.sort_index(),label='Vantaa Cycling by date')
van_by_date_max=hel_van_1516_by_date.Van_Sum.max()

ax.legend(title=('Figure 1: Helsinki and Vantaa daily cycling data on 2015 - 2016'))
l = plt.axhline(linewidth=1, color='black')
l = plt.axhline(y=hel_by_date_max, linewidth=0.5)
l = plt.axhline(y=van_by_date_max, linewidth=0.5)
```

```python
#Helsinki, Vantaa total cycling data by month on year 2015-2016

by_month=pd.DatetimeIndex(hel_van_1516.Date).to_period("M")
hel_van_1516_by_month=hel_van_1516.groupby(by_month).sum()
hel_van_1516_by_month
```

```python

df2=pd.DataFrame(hel_van_1516_by_month,index=hel_van_1516_by_month.index,columns=['Hel_Sum', 'Van_Sum']);
df2=df2.plot(kind='bar',stacked=True, figsize=(20,6),title=('Figure 2: Helsinki and Vantaa Cycling data by month'),rot=30,fontsize=10);
l = plt.axhline(y=0, linewidth=1, color='black')
```

```python
#Helsinki, Vantaa total cycling data by week on year 2015-2016
by_week=pd.DatetimeIndex(hel_van_1516.Date).to_period("W")
hel_van_1516_by_week=hel_van_1516.groupby(by_week).sum()
hel_van_1516_by_week
```

```python
df3=pd.DataFrame(hel_van_1516_by_week,index=hel_van_1516_by_week.index,columns=['Hel_Sum', 'Van_Sum']);
df3=df3.plot(kind='bar',stacked=True, figsize=(20,6),title=('Figure 3: Helsinki and Vantaa Cycling data by Week'),rot=90,fontsize=8);
l = plt.axhline(y=0, linewidth=1, color='black')
```

```python
#Helsinki, Vantaa total cycling data by quarter on year 2015-2016
by_quarter=pd.DatetimeIndex(hel_van_1516.Date).to_period("Q")
by_month=pd.DatetimeIndex(hel_van_1516.Date).to_period("M")
hel_van_1516_by_quarter_month=hel_van_1516.groupby([by_quarter,by_month]).agg({'Hel_Sum': 'sum', 'Van_Sum': 'sum'})
hel_van_1516_by_quarter_month
```

```python
df4=pd.DataFrame(hel_van_1516_by_quarter_month,index=hel_van_1516_by_quarter_month.index,columns=['Hel_Sum', 'Van_Sum']);
df4=df4.plot(kind='line',stacked=False, figsize=(20,6),title=('Figure 4: Helsinki and Vantaa Cycling data by Quarter Month'),rot=0,fontsize=10);
l = plt.axhline(y=0, linewidth=1, color='black')
```

```python
#Helsinki, Vantaa total cycling data by weekdays on timeframe

hel_van_1516_by_time = hel_van_1516.groupby(['Weekday','Time']).agg({'Hel_Sum': 'sum', 'Van_Sum': 'sum'})
hel_van_1516_by_time
```

```python
df5=pd.DataFrame(hel_van_1516_by_time,index=hel_van_1516_by_time.index,columns=['Hel_Sum', 'Van_Sum']);
df5=df5.plot(kind='line',stacked=False, figsize=(20,6),title=('Figure 5: Helsinki and Vantaa Cycling data by Weekday by time'),rot=0,fontsize=10);
l = plt.axhline(y=0, linewidth=1, color='black')
```

```python
by_month=pd.DatetimeIndex(hel_van_1516.Date).to_period("M")
hel_van_1516_by_month_weekday=hel_van_1516.groupby([by_month,'Weekday']).agg({'Hel_Sum': 'sum', 'Van_Sum': 'sum'})
hel_van_1516_by_month_weekday
```
