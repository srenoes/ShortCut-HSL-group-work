## Merge datasets together

```python
# lets merge case_new and case_death
frames = [CityBikes_04, CityBikes_05, CityBikes_06, CityBikes_07, CityBikes_08, CityBikes_09, CityBikes_10]
CityBike_all = pd.concat(frames)
CityBike_all
```

### from this onwards it is up to you if you want to use these

```python
# split date and time to two different columns, this command change value to string
CityBike_all.insert(1, 'Departure, time', CityBike_all.Departure.dt.strftime('%H:%M'))
CityBike_all.Departure = CityBike_all.Departure.dt.strftime('%Y-%m-%d')
```

```python
# split date and time to two different columns, this command change value to string
CityBike_all.insert(3, 'Return, time', CityBike_all.Return.dt.strftime('%H:%M'))
CityBike_all.Return = CityBike_all.Return.dt.strftime('%Y-%m-%d')
```
