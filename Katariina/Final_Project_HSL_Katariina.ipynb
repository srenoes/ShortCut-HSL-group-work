```python
#
# Chech file name before execute this command, just in case
#
HKI_bikes = pd.read_excel('Helsingin_pyorailijamaarat.xlsx')

# ESPOO_bikes = pd.read_excel('Espoon_pyorailijamaarat.xlsx')
# ESPOO_bikestation = pd.read_excel('Espoo_pyoralaskentapisteiden_sijainnit.xlsx')

# VANTAA_bikes = pd.read_excel('Vantaan_pyorailijamaarat.xlsx')
# VANTAA_bikestation = pd.read_excel('Vantaa_pyoralaskentapisteiden_sijainnit.xlsx')

# bikes = pd.read_csv('2019-04.csv')
```

```python
# print out dataset
HKI_bikes
```

```python
# rename columns' column names, change ä -> a and ö -> o
HKI_bikes = HKI_bikes.rename({'Päivämäärä': 'Paivamaara', 'Eteläesplanadi': 'Etelaesplanadi', 'Kaisaniemi/Eläintarhanlahti': 'Kaisaniemi/Elaintarhanlahti'}, axis=1)
HKI_bikes = HKI_bikes.rename({'Käpylä, Pohjoisbaana': 'Kapyla, Pohjoisbaana', 'Lauttasaaren silta eteläpuoli': 'Lauttasaaren silta etelapuoli'}, axis=1)
HKI_bikes = HKI_bikes.rename({'Munkkiniemen silta eteläpuoli': 'Munkkiniemen silta etelapuoli', 'Pitkäsilta itäpuoli': 'Pitkasilta itapuoli', 'Pitkäsilta länsipuoli': 'Pitkasilta lansipuoli'}, axis=1)
HKI_bikes
```

```python
#
# replace finnish word with english words due to datetime convertion
#
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('tammi', 'Jan')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('helmi', 'Feb')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('maalis', 'Mar')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('huhti', 'Apr')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('touko', 'May')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('kesä', 'Jun')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('heinä', 'Jul')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('elo', 'Aug')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('syys', 'Sep')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('loka', 'Oct')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('marras', 'Nov')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('joulu', 'Dec')

HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('ma', 'Mon')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('ti', 'Tue')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('ke', 'Wed')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('to', 'Thu')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('pe', 'Fri')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('la', 'Sat')
HKI_bikes['Paivamaara'] = HKI_bikes.Paivamaara.str.replace('su', 'Sun')
# print put result
HKI_bikes
```

```python
# Change date and time to datatime type
HKI_bikes.Paivamaara = pd.to_datetime(HKI_bikes.Paivamaara, format='%a %d %b %Y %H:%M', errors='coerce')
```
