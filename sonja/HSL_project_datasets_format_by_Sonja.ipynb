```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url_ka_hel='https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/Katariina/HelsinkiBikers_cleanDataset.csv'

#url_cycling_espoo = 'https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/Katariina/EspooBikers_cleanDataset.csv'

url_hel_weather= 'https://raw.githubusercontent.com/srenoes/ShortCut-HSL-group-work/master/csv-2e095d1c-25fe-44a6-9597-23e9bd2ba4ec.csv'
```

```python
apps2 = pd.read_csv(url_ka_hel,index_col=[1,2])
apps2
```

```python

apps2.index = apps2.index.map(' '.join)
apps2.index=pd.to_datetime(apps2.index,errors='coerce')

apps_2018=apps2.loc['2018-01-01':'2018-12-31']
apps_2018
```

```python
apps_hel_weather = pd.read_csv(url_hel_weather,index_col=[0,1,2,3])
apps_hel_weather.index=apps_hel_weather.index.map(str)
apps_hel_weather.index = apps_hel_weather.index.map(''.join)
apps_hel_weather.index=apps_hel_weather.index.map(lambda x: x.replace(', ','/'))
apps_hel_weather.index=apps_hel_weather.index.map(lambda x: x.replace('(',''))
apps_hel_weather.index=apps_hel_weather.index.map(lambda x: x.replace(')',''))
apps_hel_weather.index=pd.to_datetime(apps_hel_weather.index,errors='coerce')

apps_hel_weather
```
